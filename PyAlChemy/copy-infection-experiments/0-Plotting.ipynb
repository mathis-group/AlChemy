```r
survival.df <- all.df %>% 
                mutate(alive = count > 1) %>%
                group_by(time_step, copy_number) %>%
                summarise(ave_alive = mean(alive),
                          count = n())
survival.df <- survival.df %>%
                mutate(std_alive = sqrt(ave_alive*(1-ave_alive)/count)) %>%
                mutate(replacement = 100*copy_number/1000) %>%
                mutate(replacement_factor = as.factor(paste(replacement, "%"))) %>%
                filter(replacement >= 50)
```

```r
scientific_10 = function(x) {
  ifelse(
    x==0, "0",
    parse(text = sub("e[+]?", " %*% 10^", scales::scientific_format()(x)))
  )
} 

p <- ggplot(survival.df) +
        geom_line(aes(x= time_step*10000,
                      y = ave_alive,
                      color=replacement_factor)) +
        geom_ribbon(aes(x= time_step*10000,
                      ymin = ave_alive - std_alive,
                      ymax = ave_alive + std_alive,
                      fill=replacement_factor), alpha = 0.2) +
        labs(x = "Time (Collision Number)", y = "Organizations Surviving", color = "Identity", fill = "Identity") +
        scale_x_continuous(expand = c(0.01,0.01), label=scientific_10, breaks= c(5e5, 1e6)) +
        scale_y_continuous(expand = c(0,0,0.01,0.01)) +
        theme(plot.margin = margin(1, , 1, 1, "mm")) +
        # scale_y_log10() +
        theme_bw() + 
        scale_color_viridis_d(option = "H") +
        scale_fill_viridis_d(option = "H")
p
ggsave("L0_surivival_rate_v2.svg", p, width = 183*1/2, height = 247/4, units = "mm")
```
